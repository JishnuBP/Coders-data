In pandas we have different types of str methods and some of them are as follows

Certainly! When working with strings in pandas, you can use various string methods provided by the Series.str accessor. These methods are similar to Python's built-in string methods but are applied to each element of a pandas Series.

Here are some commonly used string methods in pandas:

1. **str.upper()**: Convert strings in the Series/Index to uppercase.
   
   ```python
   df['column'].str.upper()
   ```

2. **str.lower()**: Convert strings in the Series/Index to lowercase.
   
   ```python
   df['column'].str.lower()
   ```

3. **str.strip()**: Remove leading and trailing whitespaces from strings.
   
   ```python
   df['column'].str.strip()
   ```

4. **str.replace()**: Replace occurrences of a pattern with another string.
   
   ```python
   df['column'].str.replace('old_pattern', 'new_pattern')  ## df['column'].replace({'Yes': True, 'No' : False})
   ```

5. **str.contains()**: Check if a pattern or regex is contained within the strings.
   
   ```python
   df['column'].str.contains('pattern')
   ```

6. **str.split()**: Split strings on specified delimiter.
   
   ```python
   df['column'].str.split(',')
   ```

7. **str.join()**: Join strings in each element of the Series with specified separator.
   
   ```python
   df['column'].str.join('-')
   ```

8. **str.startswith()**: Check if strings in the Series/Index start with the specified string.
   
   ```python
   df['column'].str.startswith('prefix')
   ```

9. **str.endswith()**: Check if strings in the Series/Index end with the specified string.
   
   ```python
   df['column'].str.endswith('suffix')
   ```

10. **str.extract()**: Extract matched groups from the strings using a regex pattern.
   
    ```python
    df['column'].str.extract('(\d+)')
    ```

These are just a few examples of string methods available in pandas. You can find more methods and detailed explanations in the pandas documentation: [String handling documentation](https://pandas.pydata.org/pandas-docs/stable/user_guide/text.html)
